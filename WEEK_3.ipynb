```python
import pandas as pd
from scipy import stats
df=pd.read_excel("/content/customers-1001.xlsx")
t_statistic, p_value = stats.ttest_ind(df['Index'], df['age'])
print("t-test")
print("T_statistic:",t_statistic)
print("p_value:",p_value)
```

```python
import pandas as pd
from scipy import stats
df=pd.read_excel("/content/customers-1001.xlsx")
h_statistic, p_value = stats.kruskal(df['Index'], df['age'])
print("kruskal-wallis test")
print("H_statistic:",h_statistic)
print("p_value:",p_value)
```

```python
import pandas as pd
from scipy import stats
df=pd.read_excel("/content/customers-1001.xlsx")
u_statistic, p_value = stats.mannwhitneyu(df['Index'], df['age'])
print("wilcoxon rank-sum test")
print("U_statistic:",u_statistic)
print("p_value:",p_value)
```

```python
import pandas as pd
from scipy import stats
df=pd.read_excel("/content/customers-1001.xlsx")
f_statistic, p_value = stats.f_oneway(df['Index'], df['age'])
print("anova")
print("F_statistic:",f_statistic)
print("p_value:",p_value)
```

```python
import pandas as pd
from scipy import stats
df=pd.read_excel("/content/customers-1001.xlsx")
chi_statistic, p_value, dof, expected = stats.chi2_contingency(df['Index'], df['age'])
print("chi_square test")
print("chi_statistic:",chi_statistic)
print("p_value:",p_value)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("/content/customers-1001.xlsx")

# Calculate the correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Use seaborn to create the heatmap
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
import pandas as pd
from scipy.stats import norm
Data =pd.read_excel('/content/customers-1001.xlsx')
Mean_Sales=Data['Index'].mean()
Std_Sales=Data['Index'].std()
Value=50000
Z_score=(Value-Mean_Sales)/Std_Sales
Probability=1-norm.cdf(Z_score)
print(f'proability Index greater than ${Value}:{Probability:2%}')
```
