```python
import pandas as pd
df=pd.read_csv('/content/student_marks(01).csv')
upper_cap=1000
lower_cap=200
df_cliped=df.clip(lower=lower_cap,upper=upper_cap)
print(df_cliped)
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/student_marks(01).csv')
df_log=np.log(df)
print(df_log)
```

```python
import pandas as pd
df=pd.read_csv('/content/student_marks(01).csv')
outliers=[1000]
df_cleaned=df[~df.isin(outliers)]
print(df_cleaned)
```

```python
import pandas as pd
df=pd.read_csv('/content/student_marks(01).csv')
min_rolling=df.rolling(window=2).min()
std_rolling=df.rolling(window=2).std()
outliers=df[(df-min_rolling)>1*std_rolling]
print(outliers)
```

```python
import pandas as pd
df=pd.read_csv('/content/student_marks(01).csv')
z_score=(df-df.mean())/df.std()
print(z_score)
```
